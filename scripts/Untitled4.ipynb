```python
import tensorflow as tf
from tensorflow.keras import layers, models

inputs = tf.keras.Input(shape=(50, 3))

x = layers.Conv1D(64, kernel_size=3, activation='relu')(inputs)
x = layers.MaxPooling1D(pool_size=2)(x)
x = layers.Conv1D(128, kernel_size=3, activation='relu')(x)
x = layers.GlobalAveragePooling1D()(x)

# Attention mechanism
attention = layers.Dense(128, activation='tanh')(x)
attention = layers.Dense(1, activation='softmax')(attention)
x = layers.multiply([x, attention])

outputs = layers.Dense(len(np.unique(y_train)), activation='softmax')(x)

model = models.Model(inputs, outputs)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Flatten labels if they are nested arrays
y_train = np.ravel(y_train)
y_val   = np.ravel(y_val)
y_test  = np.ravel(y_test)

# Encode string labels into integers
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_val   = encoder.transform(y_val)
y_test  = encoder.transform(y_test)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
import numpy as np

# If y_train was expanded into sequences, flatten back to match samples
y_train = np.ravel(y_train)[:len(X_train)]
y_val   = np.ravel(y_val)[:len(X_val)]
y_test  = np.ravel(y_test)[:len(X_test)]

print("X_train:", X_train.shape)
print("y_train:", y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Clear session
tf.keras.backend.clear_session()

# Number of classes from your labels
num_classes = len(np.unique(y_train))

# Define model
inputs = layers.Input(shape=(50, 3))

x = layers.Conv1D(64, 3, activation='relu')(inputs)
x = layers.MaxPooling1D(2)(x)
x = layers.Conv1D(128, 3, activation='relu')(x)
x = layers.GlobalAveragePooling1D()(x)

# Attention-like dense layer
attention = layers.Dense(128, activation='sigmoid')(x)
x = layers.Multiply()([x, attention])

# Final output layer with correct number of classes
outputs = layers.Dense(num_classes, activation='softmax')(x)

model = models.Model(inputs=inputs, outputs=outputs)

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```
