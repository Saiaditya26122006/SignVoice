```python
import os
from PIL import Image

corrupt = []
dataset_path = "/content/dataset/isl_dataset"   # ✅ corrected path

for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    for img_file in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_file)
        try:
            img = Image.open(img_path)
            img.verify()
        except:
            corrupt.append(img_path)

print(f"Corrupt files found: {len(corrupt)}")
print(corrupt)

# ✅ Remove corrupted files
import shutil

for bad_file in corrupt:
    try:
        if os.path.isdir(bad_file):
            shutil.rmtree(bad_file)   # remove folder
        else:
            os.remove(bad_file)       # remove file
        print(f"Removed: {bad_file}")
    except Exception as e:
        print(f"Could not remove: {bad_file}, reason: {e}")

```

```python
import os
from PIL import Image

dataset_path = "/content/dataset/isl_dataset"

# Dictionary to store counts of each mode
modes_count = {}

# Dictionary to store per-folder mode distribution
folder_modes = {}

for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    if os.path.isdir(folder_path):
        folder_modes[folder] = {}
        for img_file in os.listdir(folder_path):
            img_path = os.path.join(folder_path, img_file)
            try:
                img = Image.open(img_path)
                mode = img.mode

                # Update global counts
                modes_count[mode] = modes_count.get(mode, 0) + 1

                # Update per-folder counts
                folder_modes[folder][mode] = folder_modes[folder].get(mode, 0) + 1

            except Exception as e:
                print(f"Error reading {img_path}: {e}")

# Print global summary
print("=== Global Image Mode Counts ===")
for mode, count in modes_count.items():
    print(f"{mode}: {count}")

# Print per-folder summary
print("\n=== Per-Folder Image Mode Counts ===")
for folder, modes in folder_modes.items():
    print(f"{folder}:")
    for mode, count in modes.items():
        print(f"  {mode}: {count}")
```

```python
import os, random, shutil

dataset_path = "/content/drive/MyDrive/isl_dataset"
output_path = "/content/drive/MyDrive/isl_dataset_split"

# Collect all files across all classes
all_files = []
for folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, folder)
    if os.path.isdir(folder_path):
        for f in os.listdir(folder_path):
            all_files.append((folder, os.path.join(folder_path, f)))

# Shuffle once globally
random.shuffle(all_files)

# Compute exact split sizes
n_total = len(all_files)
n_train = int(0.8 * n_total)
n_val   = int(0.1 * n_total)
n_test  = n_total - n_train - n_val

train_files = all_files[:n_train]
val_files   = all_files[n_train:n_train+n_val]
test_files  = all_files[n_train+n_val:]

# Helper to copy files into split folders
def copy_files(split_name, split_files):
    for folder, fpath in split_files:
        split_folder = os.path.join(output_path, split_name, folder)
        os.makedirs(split_folder, exist_ok=True)
        shutil.copy(fpath, os.path.join(split_folder, os.path.basename(fpath)))

# Perform the copy
copy_files("train", train_files)
copy_files("val", val_files)
copy_files("test", test_files)
```

```python
from PIL import Image
import os

def check_corrupted(path):
    corrupted = []
    for folder in os.listdir(path):
        folder_path = os.path.join(path, folder)
        if os.path.isdir(folder_path):
            for img_file in os.listdir(folder_path):
                img_path = os.path.join(folder_path, img_file)
                try:
                    img = Image.open(img_path)
                    img.verify()  # verify image integrity
                except Exception:
                    corrupted.append(img_path)
    return corrupted
```

```python
bad_train = check_corrupted(os.path.join(output_path,"train"))
print("Train corrupted:", len(bad_train))

bad_val = check_corrupted(os.path.join(output_path,"val"))
print("Val corrupted:", len(bad_val))

bad_test = check_corrupted(os.path.join(output_path,"test"))
print("Test corrupted:", len(bad_test))
```

```python
import os

def remove_corrupted(corrupted_list):
    for img_path in corrupted_list:
        try:
            os.remove(img_path)
            print(f"Removed: {img_path}")
        except Exception as e:
            print(f"Could not remove {img_path}: {e}")

# Remove corrupted files from each split
remove_corrupted(bad_train)
remove_corrupted(bad_val)
remove_corrupted(bad_test)
```
